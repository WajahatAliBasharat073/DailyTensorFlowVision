```python
params=tf.constant([[1,2,3,4],
                    [6,7,8,9],
                    [10,11,12,13]])
# Define the model
indices=tf.constant([2,0])
print(tf.gather(params, indices, axis=0))
```

```python
params = tf.constant([[1, 2, 3, 4],
                      [6, 7, 8, 9],
                      [10, 11, 12, 13]])
# Define the model
indices = tf.constant([2, 0])
print(tf.gather(params, indices, axis=1))
```

```python
params = tf.constant([[1, 2, 3, 4],
                      [6, 7, 8, 9],
                      [10, 11, 12, 13]])
# Define the model
indices = tf.constant([[2, 0],[1,1]])
print(tf.gather_nd(params, indices))
print()
print()
print(tf.gather_nd(params, indices))
print()
print(tf.gather(params, indices,axis=0))
```
